```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', None)
```

```python
path_artist = "data/Artists.csv"
path_artwork = "data/Artworks.csv"
```

```python
artist_df = pd.read_csv(path_artist)
artwork_df = pd.read_csv(path_artwork)
```

```python
artwork_df["ConstituentID"] = artwork_df["ConstituentID"].str.split()
```

```python
artwork_df = artwork_df.explode("ConstituentID").reset_index().drop("index", axis=1)
```

```python
artwork_df["ConstituentID"] = artwork_df["ConstituentID"].str.strip(",")
```

```python
artwork_df["ConstituentID"] = artwork_df["ConstituentID"].astype("float")
```

```python
artwork_data = artwork_df.merge(artist_df, left_on="ConstituentID", right_on="ConstituentID", how="left")
```
