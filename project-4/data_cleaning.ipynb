ConstituentID in artwork_df is an object datatype. To initiate a merge with artist_df, the datatype has to be changed into an integer or a float.

```python
# checking the content of the column
artwork_df["ConstituentID"].unique().tolist()
```

There seems to be multiple ConstituentID for a single artwork. It could mean that multiple artist worked on it. 
Pandas explode method will be used to split the IDs.


```python
# splitting the constituentID
artwork_df["ConstituentID"] = artwork_df["ConstituentID"].str.split(",")
artwork_df["ConstituentID"]
```

```python
# explode the artwork and reset the index
artwork_df = artwork_df.explode("ConstituentID").reset_index().drop("index", axis=1)
```

```python
# convert the datatype from object into float
artwork_df["ConstituentID"] = artwork_df["ConstituentID"].astype("float")

# print the datatype
artwork_df["ConstituentID"].dtype
```

```python
# merge the artist data into the artwork data using left join
artwork_data = artist_df.merge(artwork_df, left_on="ConstituentID", right_on="ConstituentID", how="left", suffixes=(None, '_y'))
```

```python
def extract_year(row):
    """
    function to extract date from the data column
    """
    if type(row) != float:
        row = re.findall('\d{4}', row)
        
    # get the max year from the list of year
        max = 0
        for idx in row:
            if int(idx) > max:
                max = int(idx)
        row = f"{max:.0f}" # format the date into integer
    return row
```

```python
artwork_data["completedDate"] = artwork_data["Date"].apply(extract_year)
artwork_data["completedDate"].head()
```
